```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the datasets
print("Reading the data . . .")
covid19_data = pd.read_csv('datasets/covid19_Confirmed_dataset.csv') # Source: John Hopkins University
life_factors = pd.read_csv('datasets/worldwide_happiness_report.csv') # Source: United Nations
print("Data is imported!")
```

```python
covid19_data.head(10)
```

```python
covid19_data.describe()
```

```python
covid19_data.shape
```

```python
#Removing the unnecessary columns from our dataset
covid19_data.drop(labels = ["Lat","Long"],axis = 1,inplace = True)
covid19_data.head(10)
```

```python
covid19_data_agg = covid19_data.groupby(by = "Country/Region").sum()
covid19_data_agg.head(10)
```

```python
covid19_data_agg.shape
```

```python
#Infection rate (or daily infected cases)
covid19_infection_rate = covid19_data_agg.aggregate(func = 'diff',axis = 1)
```

```python
covid19_infection_rate.head(10)
```

```python
#Maximum infection rate of every country
max_infection_rate = covid19_infection_rate.max(axis = 1)
```

```python
covid19_data_agg["Max Infection Rate"] = max_infection_rate
```

```python
covid19_data_agg.head(10)
```

```python
covid19_final_data = pd.DataFrame(covid19_data_agg["Max Infection Rate"])
```

```python
covid19_final_data.head(10)
```

```python
life_factors.head(10)
```

```python
#Removing the unnecessary columns from the dataset life_factors
life_factors.drop(['Overall rank','Score','Generosity','Perceptions of corruption'],axis = 1,inplace = True)
```

```python
life_factors.head(10)
```

```python
life_factors.set_index(keys = 'Country or region',inplace = True)
```

```python
#Concatenating the two datasets: covid19_final_data & life_factors
data = covid19_final_data.join(life_factors,how = 'inner')
```
